```python
# Dependencies
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# list of TV show titles to query
tv_shows = ["Suits", "Bobs Burgers", "What We Do In The Shadows","The Office",
            "Always Sunny in Philadelphia","House of the Dragon","Family Guy",
            "Game of Thrones"]

# TV Maze show search base URL
base_url = "http://api.tvmaze.com/search/shows?q="

# set up lists to hold response data for name and rating
titles = []
ratings = []

# loop through TV show titles, make requests and parse
for show in tv_shows:
    target_url = base_url + show
    response = requests.get(target_url).json()
    titles.append(response[0]['show']['name'])
    ratings.append(response[0]['show']['rating']['average'])
```

```python
# create DataFrame
shows_df = pd.DataFrame({
    "title": titles,
    "rating": ratings
})

shows_df
```

```python
# Plot the data
shows_df.plot('title', 'rating', kind='bar', figsize=(10,6), color='b', alpha=.8, rot=45)
plt.title("Average Zoe Show Rating")
plt.ylim(6,10)
```
